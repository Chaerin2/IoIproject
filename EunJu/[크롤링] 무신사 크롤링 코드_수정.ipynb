```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time

from webdriver_manager.chrome import ChromeDriverManager
driver=webdriver.Chrome()

import pandas as pd

import re

import requests
from bs4 import BeautifulSoup

url = 'https://www.musinsa.com/app/goods/1495284'

driver.get(url)
driver.implicitly_wait(3)
time.sleep(5)

html_source = driver.page_source
soup = BeautifulSoup(html_source, 'html.parser')

title = soup.select('#page_product_detail > div.right_area.page_detail_product > div.right_contents.section_product_summary > span > em')
name = title[0].text

img = soup.select('#detail_bigimg > div.product-img > img')
imglink = img[0].get('src')

li = soup.select('#product_size_recommend > ul > li')
df = pd.DataFrame(columns=['제품명','성별','키','몸무게','핏','이미지링크','구매사이즈'])

for i in range(0, len(li)):
    try:
        size_content = li[i].find('p', {'class', 'size_content'})
        
        info = size_content.text.replace('\n', '').replace('\t', '').strip()
        fit = info.split('기준')[1].strip().split(' ')[0]
        
        pattern_gender = r'(?<=\()\w+(?=\s+\d+cm/\d+kg\))' # 성별
        pattern_height = r'\d+cm' # 키
        pattern_kg = r'\d+kg' # 몸무게

        pattern_size = r'\D\s+Size' # 사이즈
        pattern_fit = info.split(' ')

        # 정규표현식 매칭
        gender = re.search(pattern_gender, info).group()
        height = re.search(pattern_height, info).group()
        kg = re.search(pattern_kg, info).group()
        size = re.search(pattern_size, info).group().split(' ')[0]
        
        df = df.append({'제품명': name,
                        '성별': gender,
                        '키': height,
                        '몸무게': kg,
                        '핏': fit,
                        '이미지링크': imglink,
                        '구매사이즈': size,
                       }, ignore_index = True)
    except:
        pass

df.to_csv('9_슬랙스.csv',encoding='utf-8-sig')
```
