```python
import pandas as pd
df = pd.read_csv('수정_clothes_pandas_.csv')
```

```python
product_name_mapping = {product_name: idx for idx, product_name in enumerate(df['제품명'].unique())}
df['제품번호'] = df['제품명'].map(product_name_mapping)
```

```python
df['사이즈_제품번호'] = df['구매사이즈'] + df['제품번호'].astype(str)
new_columns_order = ['기온', '제품번호','제품명', '성별', '키', '몸무게', '핏', '이미지링크', '구매사이즈','사이즈_제품번호']
df = df[new_columns_order]
```

```python
df.columns = ['temp','c_number','c_name','gender','height(cm)','weight(kg)', 'fit','link','size','c_result']
```
