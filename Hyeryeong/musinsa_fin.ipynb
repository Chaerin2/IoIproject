* musinsa : 원본 데이터 
* data1 : c_number(int -> object) 진행했다가 다시 복원함 : musinsa랑 동일
    * c_number를  object type으로 변환 시, 딕셔너리 key('c_number')를 통해 value('c_name') 추출할 때 문자열로 취급해야하므로 복원
* data2 : 수정 데이터 
* data3 : c_number, c_name 중복 제거
* result_dict : c_number와 c_name & link 매핑

```python
import pandas as pd
musinsa = pd.read_csv("./musinsa_data.csv")
```

```python
musinsa.head()
```

```python
musinsa.info()
```

```python
data1 = musinsa
data1
```

```python
data2 = data1.drop(['c_number', 'c_name', 'link', 'size'], axis=1)
data2
```

```python
data3 = data1.drop_duplicates(subset=['c_name', 'c_number'], keep='first')
data1.info()
```

```python
result_dict = {key: {'c_name': val['c_name'].tolist(), 'link': val['link'].tolist()} for key, val in data3.groupby('c_number')}
print(result_dict)
```
